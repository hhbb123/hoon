```python
import random
import numpy
import matplotlib.pyplot as plt
a0=[1,1,1]
a1=[2,1,3]
a2=[3,2,1]
a3=[2,2,2]
a4=[3,3,3]
A1=[a0,a1,a2,a3,a4]
B=[]
Bm2=[]
Bm3=[]
Bd2=[]
Bd3=[]

for i in range(1024):
    a=random.randrange(0,4)
    B.append(A1[a])
print(B)
    
    
```

```python
def Mendel(p,q):
    R=[p,q]
    r0=random.randrange(0,2)
    r1=random.randrange(0,2)
    r2=random.randrange(0,2)
    return [R[r0][0],R[r1][1],R[r2][2]]
print(Mendel([1,2,3],[4,5,6]))
    
```

```python
for s1 in range(1024):
    #여기서 if를 써서 자연선택을 표현하는 것이 좋을듯
    Bm2.append(Mendel(B[s1],B[s1-1]))
for s2 in range(562):
    Bm3.append(Mendel(Bm2[s2],Bm2[s2-1]))
    
Gm0=[]
Gm1=[]
Gm2=[]
for g in range(562):
    Gm0.append(Bm3[g][0])
    Gm1.append(Bm3[g][1])
    Gm2.append(Bm3[g][2])

plt.figure(1)
plt.hist(Gm0)
plt.figure(2)
plt.hist(Gm1)
plt.figure(3)
plt.hist(Gm2)
plt.figure(4)
plt.hist(Bm3)
plt.show()


    
```

```python
def Darwin(p,q):
    return [(p[0]+q[0])/2,(p[1]+q[1])/2,(p[2]+q[2])/2]
```

```python
for sd1 in range(1024):
    Bd2.append(Darwin(B[sd1],B[sd1-1]))
for sd2 in range(562):
    Bd3.append(Darwin(Bd2[sd2],Bd2[sd2-1]))

plt.hist(Bd2)
plt.show()
    
    
```

```python
import matplotlib.pyplot as plt
import random

A1=[0,0]
A2=[1,0]
A3=[0.5,(3**0.5)/2]
CA=[A1,A2,A3]
IP=[random.random(),random.random()]


def MPF(a,b):
    c=[]
    for i in range(len(a)):
        r=(a[i]+b[i])/2
        c.append(r)
    return c


IPLXR=[]
IPLYR=[]

IPLXG=[]
IPLYG=[]

IPLXB=[]
IPLYB=[]

for i in range(10000):
    RL=random.randint(0,2)
    CP=CA[RL]
    IP=MPF(IP,CP)
    if RL==0 :
        IPLXR.append(IP[0])
        IPLYR.append(IP[1])
    if RL==1:
        IPLXG.append(IP[0])
        IPLYG.append(IP[1])
    if RL==2:
        IPLXB.append(IP[0])
        IPLYB.append(IP[1])
        
plt.scatter(IPLXR,IPLYR,s=1,c='r')
plt.scatter(IPLXG,IPLYG,s=1,c='g')
plt.scatter(IPLXB,IPLYB,s=1,c='b')
plt.show()


```

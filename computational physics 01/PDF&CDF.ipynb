```python
import random
import matplotlib.pyplot as plt
import math
import numpy as np
def exponTF(t):
    A=[]
    for i in range(t):
        a=random.random()
        A.append(-math.log(1-a))
    return A
plt.hist(exponTF(100000),bins=100, density=True)

def exponTFA(t):
    A=[]
    for i in range(t):
        a=random.random()
        A.append((-math.log(1-a))*(-math.log(1-a)))
    return A


plt.hist(exponTF(100000),bins=100, density=True)

B=[]
for x in np.arange(0,10,0.1):
    c=math.exp(-x)
    B.append(c)
plt.plot(np.arange(0,10,0.1),B)

plt.show()
plt.show()
```

```python
#L=plt.hist(exponTF(100000),bins=100)
mean1=sum(exponTF(1000000))/1000000
mean2=sum(exponTFA(1000000))/1000000
vera= mean2-mean1*mean1

print(mean1)
print(vera)
#print(vera)
```

```python
import random
import matplotlib.pyplot as plt
import math
import numpy as np
def exponTF1(t):
    A=[]
    for i in range(t):
        a=random.random()
        A.append(-3*math.log(1-a))
    return A
def exponTF2(t):
    A=[]
    for i in range(t):
        a=random.random()
        A.append((-3*math.log(1-a))*(-3*math.log(1-a)))
    return A

B=[]
for x in np.arange(0,10,0.1):
    c=(1/3)*math.exp(-(1/3)*x)
    B.append(c)
plt.plot(np.arange(0,10,0.1),B)


plt.hist(exponTF1(100000),bins=100,density=True)
plt.show()
```

```python
#L=plt.hist(exponTF(100000),bins=100)
mean1=sum(exponTF1(1000000))/1000000
mean2=sum(exponTF2(1000000))/1000000
va=mean2-mean1*mean1

print(mean1)
print(va)
```

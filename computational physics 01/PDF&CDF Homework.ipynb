```python
#0.5~1 사이 균일한 PDF
import numpy as np
import matplotlib.pyplot as plt
UA=[]
for i in range(1000000):
    x=0.5*np.random.rand()+0.5 #0.5와 1사이로 변환
    UA.append(x)
plt.hist(UA,bins=100,density=True)
plt.ylabel('PDF')
plt.title('0.5~1 uniform PDF')
plt.show()
```

```python
#0.5~ 1사이 균일한 CDF 및 이론
BU=[]
for x in np.arange(0.5,1,0.0001):
    UY=2*x-1
    BU.append(UY)
plt.plot(np.arange(0.5,1,0.0001),BU)#이론값
plt.hist(UA,bins=200,density=True,cumulative=True)#시뮬레이션한 값
plt.ylabel('CDF')
plt.title('0.5~1 uniform CDF')
plt.show
```

```python
# mean=0.5 Exponential PDF
import numpy as np
import matplotlib.pyplot as plt
import math
def exPDF(t):
    EA=[]
    for i in range(t):
        ea=np.random.random()
        EA.append(-0.5*math.log(1-ea))
    return EA #변환으로 얻은값
EB=[]
for x in np.arange(0,10,0.1):
    EY=2*math.exp(-2*x)
    EB.append(EY) #수식적으로 얻은값
plt.plot(np.arange(0,10,0.1),EB)
plt.hist(exPDF(100000),bins=200,density=True)
plt.ylabel('PDF')
plt.title('mean 0.5 Exponential PDF')
plt.show()
```

```python
# mean=0.5 Exponential PDF
plt.hist(exPDF(100000),bins=200,density=True,cumulative=True)
plt.ylabel('CDF')
plt.title('mean 0.5 Exponential PDF')
plt.show()
```

```python
#mean 1 variance 3 Gausian PDF
import matplotlib.pyplot as plt
import random
import math
def gausian(sigma):
    flag=True
    if(flag):
        r=math.sqrt(-2*(sigma**2)*math.log(1-random.random()))
        theta=2*math.pi*random.random()
        reserve=r*math.cos(theta)
        flag=False
        return r*math.sin(theta)
    else:
        flag=True
        return reserve
GA=[]
for i in range(100000):
    ag=gausian(3**0.5)+1
    GA.append(ag) #시뮬레이션한 값
GB=[]
for x in np.arange(-5,5,0.01):
    GY=(1/((3**0.5)*(2*math.pi)**0.5))*math.exp((-(x-1)**2)/(2*3))
    GB.append(GY) # 이론값
plt.plot(np.arange(-5,5,0.01),GB)
plt.hist(GA,bins=100,density=True)
plt.ylabel('PDF')
plt.title('mean1 variance 3 Gausian PDF')
plt.show()
```

```python
#mean 1 variance 3 Gausian CDF
plt.hist(GA,bins=100,density=True,cumulative=True)
plt.ylabel('CDF')
plt.title('mean 1 variance 3 CDF')
plt.show()
```

```python
#r=2 medium1 cauchy PDF
import numpy
import matplotlib.pyplot as plt
import math

def couran(alpha):
    y=numpy.random.random()
    x=2*numpy.tan(math.pi*(y-alpha))+1
    return x
CA=[]
for i in range(100000):
    rc=couran(0.5)
    if -25<rc<25:
        CA.append(rc) #시뮬레이션
plt.ylabel('PDF')
plt.title('r=2 medium1 cauchy PDF')
plt.hist(CA,density=True,bins=300)
CB=[]
for x in numpy.arange(-25,25,0.01):
    CY=(1/math.pi)*(2/(((x-1)**2)+4))
    CB.append(CY)#이론값
plt.plot(numpy.arange(-25,25,0.01),CB)
plt.show()
```

```python
#r=2 medium1 cauchy CDF
plt.hist(CA,density=True,bins=300,cumulative=True)
plt.xlabel('CDF')
plt.title('r=2 medium1 cauchy CDF')
```

```python
# xm 1 alpha 2.5 Pareto PDF
import numpy as np
import matplotlib. pyplot as plt
import math

def PTran(alpha):
    xp=np.random.random()+2
    return (1/(0.6*xp-1))**alpha
PA=[]
print(PTran(1.5))
for i in range(10000):
    pr=PTran(1.5)
    PA.append(pr)
PB=[]
for x in np.arange(2,5,0.01):
    yb=2.5*((1/x)**2.5)
    PB.append(yb)
plt.ylabel('PDF')
plt.title('xm 1 alpha 2.5 Pareto PDF')
plt.hist(PA,density=True,bins=200)
plt.plot(np.arange(2,5,0.01),PB)
plt.show()
```

```python
import random
import math
import matplotlib .pyplot as plt
```

```python
s1=0
s2=0
x=[]
y=[]
for i in range(10000):
    a=random.random()
    b=a*2*math.pi
    s1 += math.cos(b)
    s2 += math.sin(b)
    x.append(s1)
    y.append(s2)
x_last=x[-1]
y_last=y[-1]
plt.plot(x,y)
plt.plot(x_last,y_last, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import random
import matplotlib .pyplot as plt
import math

def random_Final(n):
    s1=0
    s2=0
    for i in range(n):
        a=random.random()
        b=2*a*math.pi
        s1 += math.cos(b)
        s2 += math.sin(b)
        s3=(s1**2+s2**2)**0.5
    return [s1,s2,s3]
F1=[]
F2=[]
F3=[]
for i in range(1000):
    F1.append(random_Final(10000)[0])
    F2.append(random_Final(10000)[1])
    F3.append(random_Final(10000)[2])
    
```

```python
plt.hist(F1,bins=200)
plt.title('x-axis distance')
plt.xlabel('distance')
plt.ylabel('Number')
plt.show()
```

```python
def list_distance(a,b):
    c=[]
    for i in range(0,len(a)):
        distance=a[i]*a[i]+b[i]*b[i]
        c.append(distance)
    return c

    
```

```python
import numpy
def list_random_distance(n):
    s1=0
    s2=0
    x=[]
    y=[]
    for i in range(n):
        a=random.random()
        b=a*2*math.pi
        s1 += math.cos(b)
        s2 += math.sin(b)
        x.append(s1)
        y.append(s2)
    r=list_distance(x,y)
    return r
time=[]
mean1=[]
for i in range(1000):
    for t in range(10000):
        time.append(list_random_distance(10000)[t])
    mean1.append(numpy.mean(time))
plt.plot(mean1)
plt.show()

    
    
```

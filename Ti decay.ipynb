```python
#방사능 붕괴
import random
import matplotlib. pyplot as plt
import math
tau=3.053*60
h=1.0
p=1.0-math.pow(2,-h/tau) #붕괴 확률
tN=0
NTi=1000

list_NTi=[]
list_Npb=[]
list_N=[]
Npb=0
for tN in range(0,1000): #이론값
    N=NTi*math.pow(2,-tN/tau)
    list_N.append(N)
    
for t in range(0,1000):   #실제로 카운트
    decay=0
    for i in range(0,NTi):
        if random.random()<p :
            decay=decay+1
    NTi -= decay
    Npb += decay
    list_NTi.append(NTi)
    list_Npb.append(Npb)
plt.plot(list_NTi)
plt.plot(list_Npb)
plt.plot(list_N)
plt.show()
```

```python
from pylab import xlabel, ylabel, title, plot, show, legend
import random

def proportion(t, tau):
    return 1-2**(-1*t/tau)

tau_Bi=46*60
tau_Tl=2.2*60
tau_Pb=3.3*60*60
h=1.0
p_Bi=proportion(h,tau_Bi)
p_Tl=proportion(h,tau_Tl)
p_Pb=proportion(h,tau_Pb)
NBi=10000
NTl=0
NPb=0
Nstable=0

NBi_list=list()
NTl_list=list()
NPb_list=list()
Nstable_list=list()

for t in range(100000):
    decay_Bi=0
    decay_Tl=0
    decay_Pb=0
    
    for i in range(NPb):
        if random.random()<p_Pb:
            decay_Pb+=1
            
   
    for i in range(NTl):
        if random.random()<p_Tl:
            decay_Tl+=1
            
    
    for i in range(NBi):
        if  random.random()<p_Bi:
            decay_Bi+=1
    
    #Bi가 2.09% 확률로 Tl로 붕괴. 아니면 Pb로 붕괴
    for i in range(decay_Bi):
        if random.random()<0.0209:
            NTl+=1
        else:
            NPb+=1
    
    
    NBi-=decay_Bi
    NTl-=decay_Tl
    NPb-=decay_Pb
    
    
    NPb+=decay_Tl
    
    Nstable+=decay_Pb
    
    NPb_list.append(NPb)
    NTl_list.append(NTl)
    NBi_list.append(NBi)
    Nstable_list.append(Nstable)

plot(NBi_list,label='Bi')
plot(NTl_list,label='Tl')
plot(NPb_list,label='Pb')
plot(Nstable_list,label='Stable')
xlabel('sec')
ylabel('Number')

show()
```

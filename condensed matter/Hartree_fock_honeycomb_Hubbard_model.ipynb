```python
import numpy as np
import random
import pylab as plt
```

```python
# Hartree fock armchair honeycomb hamiltonian

def Hamiltonian_up(t,U,n_down_set,k): #끝 부분에 hopping이 안됨
  Hamiltonian_up_matrix=np.zeros([len(n_down_set),len(n_down_set)])
  for i in range(0,len(n_down_set)-8):
    if i%8==0:
      Hamiltonian_up_matrix[i][i-7]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i+1]=-t*np.cos(k)
      
      Hamiltonian_up_matrix[i-7][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i+1][i]=-t*np.cos(k)

      Hamiltonian_up_matrix[i][i]=U*n_down_set[i]-0.5*U
    
    elif i%8==1:
      Hamiltonian_up_matrix[i][i+7]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i+1]=-t*np.cos(k)

      Hamiltonian_up_matrix[i+7][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i+1][i]=-t*np.cos(k)

      Hamiltonian_up_matrix[i][i]=U*n_down_set[i]-0.5*U

    elif i%8==2:
      Hamiltonian_up_matrix[i][i+9]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i+1]=-t*np.cos(k)

      Hamiltonian_up_matrix[i+9][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i+1][i]=-t*np.cos(k)

      Hamiltonian_up_matrix[i][i]=U*n_down_set[i]-0.5*U

    elif i%8==3:
      Hamiltonian_up_matrix[i][i-9]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i+1]=-t*np.cos(k)

      Hamiltonian_up_matrix[i-9][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i+1][i]=-t*np.cos(k)

      Hamiltonian_up_matrix[i][i]=U*n_down_set[i]-0.5*U
    
    elif i%8==4:
      Hamiltonian_up_matrix[i][i-7]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i+1]=-t*np.cos(k)

      Hamiltonian_up_matrix[i-7][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i+1][i]=-t*np.cos(k)

      Hamiltonian_up_matrix[i][i]=U*n_down_set[i]-0.5*U
    
    elif i%8==5:
      Hamiltonian_up_matrix[i][i+7]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i+1]=-t*np.cos(k)

      Hamiltonian_up_matrix[i+7][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i+1][i]=-t*np.cos(k)

      Hamiltonian_up_matrix[i][i]=U*n_down_set[i]-0.5*U

    elif i%8==6:
      Hamiltonian_up_matrix[i][i+9]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i+1]=-t*np.cos(k)

      Hamiltonian_up_matrix[i+9][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i+1][i]=-t*np.cos(k)

      Hamiltonian_up_matrix[i][i]=U*n_down_set[i]-0.5*U
    
    elif i%8==7:
      Hamiltonian_up_matrix[i][i-9]=-t*np.cos(k)
      Hamiltonian_up_matrix[i][i-1]=-t*np.cos(k)

      Hamiltonian_up_matrix[i-9][i]=-t*np.cos(k)
      Hamiltonian_up_matrix[i-1][i]=-t*np.cos(k)

      Hamiltonian_up_matrix[i][i]=U*n_down_set[i]-0.5*U

  Hamiltonian_up_matrix[len(n_down_set)-8][len(n_down_set)-8]=U*n_down_set[len(n_down_set)-8]-0.5*U
  Hamiltonian_up_matrix[len(n_down_set)-7][len(n_down_set)-7]=U*n_down_set[len(n_down_set)-7]-0.5*U
  Hamiltonian_up_matrix[len(n_down_set)-6][len(n_down_set)-6]=U*n_down_set[len(n_down_set)-6]-0.5*U
  Hamiltonian_up_matrix[len(n_down_set)-5][len(n_down_set)-5]=U*n_down_set[len(n_down_set)-5]-0.5*U
  Hamiltonian_up_matrix[len(n_down_set)-4][len(n_down_set)-4]=U*n_down_set[len(n_down_set)-4]-0.5*U
  Hamiltonian_up_matrix[len(n_down_set)-3][len(n_down_set)-3]=U*n_down_set[len(n_down_set)-3]-0.5*U
  Hamiltonian_up_matrix[len(n_down_set)-2][len(n_down_set)-2]=U*n_down_set[len(n_down_set)-2]-0.5*U
  Hamiltonian_up_matrix[len(n_down_set)-1][len(n_down_set)-1]=U*n_down_set[len(n_down_set)-1]-0.5*U

  Hamiltonian_up_matrix[len(n_down_set)-1][len(n_down_set)-2]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-2][len(n_down_set)-1]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-2][len(n_down_set)-3]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-3][len(n_down_set)-2]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-3][len(n_down_set)-4]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-4][len(n_down_set)-3]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-4][len(n_down_set)-5]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-5][len(n_down_set)-4]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-5][len(n_down_set)-6]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-6][len(n_down_set)-5]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-6][len(n_down_set)-7]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-7][len(n_down_set)-6]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-7][len(n_down_set)-8]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-8][len(n_down_set)-7]=-t*np.cos(k)


  Hamiltonian_up_matrix[len(n_down_set)-2][7]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-3][4]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-6][3]=-t*np.cos(k)
  Hamiltonian_up_matrix[len(n_down_set)-7][0]=-t*np.cos(k)

  Hamiltonian_up_matrix[7][len(n_down_set)-2]=-t*np.cos(k)
  Hamiltonian_up_matrix[4][len(n_down_set)-3]=-t*np.cos(k)
  Hamiltonian_up_matrix[3][len(n_down_set)-6]=-t*np.cos(k)
  Hamiltonian_up_matrix[0][len(n_down_set)-7]=-t*np.cos(k)


  return Hamiltonian_up_matrix

def Hamiltonian_down(t,U,n_up_set,k):
  Hamiltonian_down_matrix=np.zeros([len(n_up_set),len(n_up_set)])
  for i in range(0,len(n_up_set)-8):
    if i%8==0:
      Hamiltonian_down_matrix[i][i-7]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i+1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i]=U*n_up_set[i]-0.5*U

      Hamiltonian_down_matrix[i-7][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+1][i]=-t*np.cos(k)
    
    elif i%8==1:
      Hamiltonian_down_matrix[i][i+7]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i+1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+7][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i]=U*n_up_set[i]-0.5*U

    elif i%8==2:
      Hamiltonian_down_matrix[i][i+9]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i+1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+9][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i]=U*n_up_set[i]-0.5*U

    elif i%8==3:
      Hamiltonian_down_matrix[i][i-9]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i+1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i-9][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i]=U*n_up_set[i]-0.5*U
    
    elif i%8==4:
      Hamiltonian_down_matrix[i][i-7]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i+1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i-7][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i]=U*n_up_set[i]-0.5*U
    
    elif i%8==5:
      Hamiltonian_down_matrix[i][i+7]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i+1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+7][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i]=U*n_up_set[i]-0.5*U

    elif i%8==6:
      Hamiltonian_down_matrix[i][i+9]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i+1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+9][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i+1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i]=U*n_up_set[i]-0.5*U
    
    elif i%8==7:
      Hamiltonian_down_matrix[i][i-9]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i-1]=-t*np.cos(k)
      Hamiltonian_down_matrix[i-9][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i-1][i]=-t*np.cos(k)
      Hamiltonian_down_matrix[i][i]=U*n_up_set[i]-0.5*U

  Hamiltonian_down_matrix[len(n_up_set)-8][len(n_up_set)-8]=U*n_up_set[len(n_up_set)-8]-0.5*U
  Hamiltonian_down_matrix[len(n_up_set)-7][len(n_up_set)-7]=U*n_up_set[len(n_up_set)-7]-0.5*U
  Hamiltonian_down_matrix[len(n_up_set)-6][len(n_up_set)-6]=U*n_up_set[len(n_up_set)-6]-0.5*U
  Hamiltonian_down_matrix[len(n_up_set)-5][len(n_up_set)-5]=U*n_up_set[len(n_up_set)-5]-0.5*U
  Hamiltonian_down_matrix[len(n_up_set)-4][len(n_up_set)-4]=U*n_up_set[len(n_up_set)-4]-0.5*U
  Hamiltonian_down_matrix[len(n_up_set)-3][len(n_up_set)-3]=U*n_up_set[len(n_up_set)-3]-0.5*U
  Hamiltonian_down_matrix[len(n_up_set)-2][len(n_up_set)-2]=U*n_up_set[len(n_up_set)-2]-0.5*U
  Hamiltonian_down_matrix[len(n_up_set)-1][len(n_up_set)-1]=U*n_up_set[len(n_up_set)-1]-0.5*U

  Hamiltonian_down_matrix[len(n_up_set)-1][len(n_up_set)-2]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-2][len(n_up_set)-1]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-2][len(n_up_set)-3]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-3][len(n_up_set)-2]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-3][len(n_up_set)-4]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-4][len(n_up_set)-3]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-4][len(n_up_set)-5]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-5][len(n_up_set)-4]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-5][len(n_up_set)-6]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-6][len(n_up_set)-5]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-6][len(n_up_set)-7]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-7][len(n_up_set)-6]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-7][len(n_up_set)-8]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-8][len(n_up_set)-7]=-t*np.cos(k)


  Hamiltonian_down_matrix[len(n_up_set)-2][7]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-3][4]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-6][3]=-t*np.cos(k)
  Hamiltonian_down_matrix[len(n_up_set)-7][0]=-t*np.cos(k)
  
  Hamiltonian_down_matrix[7][len(n_up_set)-2]=-t*np.cos(k)
  Hamiltonian_down_matrix[4][len(n_up_set)-3]=-t*np.cos(k)
  Hamiltonian_down_matrix[3][len(n_up_set)-6]=-t*np.cos(k)
  Hamiltonian_down_matrix[0][len(n_up_set)-7]=-t*np.cos(k)

  

  return Hamiltonian_down_matrix
```

```python
# inital value

n_up_set=np.zeros(1024)
n_down_set=np.zeros(1024)

for i in range(0,512):
  n_up_set[2*i]=   0.2
  n_down_set[2*i]=   0.8
  n_up_set[2*i+1]=   0.8
  n_down_set[2*i+1]=   0.2
```

```python
# value
t=1
U=3
```

```python
# self-contained

E_list=[]

E_prev=1
E=1000
while abs(E_prev-E)>1e-8:
  V_up,F_up=np.linalg.eig(Hamiltonian_up(t,U,n_down_set,0))
  V_up_index=np.argsort(V_up)
  V_up_new=V_up[V_up_index]
  w_up_new=F_up[:,V_up_index]
  for j in range(0,len(n_up_set)):
    n_up_set[j]=0
  for j in range(0,len(w_up_new[0])):
    for m in range(0,512):  #half-full
      n_up_set[j] += np.conjugate(w_up_new[j][m])*w_up_new[j][m]
  
  V_down,F_down=np.linalg.eig(Hamiltonian_down(t,U,n_up_set,0))
  V_down_index=np.argsort(V_down)
  V_down_new=V_down[V_down_index]
  w_down_new=F_down[:,V_down_index]
  w_down=F_down[:,np.argmin(V_down)]
  for j in range(0,len(n_down_set)):
    n_down_set[j]=0
  for j in range(0,len(w_down_new[0])):
    for m in range(0,512): #half-full
      n_down_set[j] += np.conjugate(w_down_new[j][m])*w_down_new[j][m]
  
  e_up=0
  e_down=0
  e_interaction=0
  for j in range(0,512):
    e_up += V_up_new[j]
  for j in range(0,512):
    e_down += V_down_new[j]
  for j in range(0,1024):
    e_interaction += U*n_up_set[j]*n_down_set[j]

  E_prev=E
  E=e_up+e_down-e_interaction
  E_list.append(E)

print(e_interaction)
print(e_down)
print(e_up)
print(E) 
```

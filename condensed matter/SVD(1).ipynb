```python
import numpy as np
```

```python
# random matrix
a=int(input())
b=int(input())
A=np.random.rand(a,b)
print(A)
```

```python
#SVD
U,S,Vt=np.linalg.svd(A)
print('U \n',U)
print('sigma \n',S)
print('V tau \n',Vt)
```

```python
#복원
Sm=np.zeros((a,b))
for i in range(min(a,b)):
    Sm[i][i]=S[i]
A_after=np.dot(np.dot(U,Sm),Vt)
print(A_after)
```

```python
import numpy as np
import pylab as plt

x0=0
y0=0
x_v0=5
y_v0=5
b=1
g=9.8

N=100000
tpoints=np.arange(0,N,0.01)
velocities=np.zeros((N,2))

init_postition=np.array((x0,y0))
init_velocity=np.array((x_v0,y_v0))

dt=tpoints[1]-tpoints[0]

positions=np.zeros((N,2))
velocities[0]=init_velocity

def acc(x,v):
  x_vel=v[0]
  y_vel=v[1]
  dxdt=x_vel
  dydt=y_vel

  dx_vel_dt=-b*dxdt**2
  dy_vel_dt=-g-b*dydt**2

  return np.array([dx_vel_dt,dy_vel_dt],float)

vh=np.zeros((N,2))
vh[0]=velocities[0]+acc(positions[0],velocities[0])*dt/2

acces=np.zeros((N,2))
acces[0]=acc(positions[0],velocities[0])

for i in range(N-1):
  positions[i+1]=positions[i]+dt*vh[i]
  vh[i+1]=vh[i]+dt*acc(positions[i+1],vh[i])
  velocities[i+1]=vh[i+1]-dt/2*acc(positions[i+1],velocities[i])
  if positions[i][1]<0:
    break

positions=positions[:i]
velocities=velocities[:i]
```

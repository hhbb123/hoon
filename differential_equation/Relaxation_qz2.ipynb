```python
import numpy as np
import pylab as plt
```

```python
a=0
b=10
N=100

n=(b-a)/N
tpoints=np.arange(a,b+n,n)

phi=np.zeros((N+1,N+1))
phi[0,:]=1

phi_prev=np.zeros((N+1,N+1))
phi_prev[0,:]=1

flag=True

target=1e-5
```

```python
while flag:
  flag=False
  for i in range(1,N):
    for j in range(1,N):
      phi_prev[i][j]=0.25*(phi[i+1][j]+phi[i-1][j]+phi[i][j+1]+phi[i][j-1])
      if flag==False:
        if abs(phi_prev[i][j]-phi[i][j])>target:
          flag=True
  phi=1.0*phi_prev

```

```python
plt.imshow(phi)
plt.colorbar()
```

```python
phi_ch=np.zeros((N+1,N+1))
phi_ch_prev=np.zeros((N+1,N+1))

flag=True

target=1e-5
```

```python
rho=np.zeros((N+1,N+1))
for i in range(59,79):
  for j in range(19,39):
    rho[i][j]=1

for i in range(19,39):
  for j in range(59,79):
    rho[i][j]=-1
```

```python
while flag:
  flag=False
  for i in range(1,N):
    for j in range(1,N):
      phi_ch_prev[i][j]=0.25*(phi_ch[i+1][j]+phi_ch[i-1][j]+phi_ch[i][j+1]+phi_ch[i][j-1])+(0.25*1e-4/(8.85e-12))*rho[i][j]
      if flag==False:
        if abs(phi_ch_prev[i][j]-phi_ch[i][j])>target:
          flag=True
  phi_ch=1.0*phi_ch_prev
```

```python
plt.imshow(phi_ch)
plt.colorbar()
```

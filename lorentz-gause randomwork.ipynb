```python
import numpy
import matplotlib.pyplot as plt
import math

def couran(alpha):
    y=numpy.random.random()
    x=numpy.tan(math.pi*(y-alpha))
    return x
A=[]
print(couran(0.5))
for i in range(100000):
    rc=couran(0.5)
    if -25<rc<25:
        A.append(rc)

plt.hist(A,density=True,bins=300)
plt.show()
    
```

```python
X=[]
Y=[]
sx=0
sy=0
for i in range(1000):
    a=numpy.random.random()
    b=2*math.pi*a
    r=couran(0.5)
    sx += r*math.cos(b)
    sy += r*math.sin(b)
    X.append(sx)
    Y.append(sy)
plt.plot(X,Y)
plt.show()
```
